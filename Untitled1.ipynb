```python
import pandas as pd
import numpy as np
admissions = pd.read_csv("Admission_Predict.csv")
pd.options.display.max_columns = 99
pd.options.display.width = 150
print(admissions.head())
```

```python
#The different factures enlisted are 1. GRE Scores ( out of 340 ) 2. TOEFL Scores ( out of 120 ) 3. University Rating ( out of 5 )
#4. Statement of Purpose and Letter of Recommendation Strength ( out of 5 ) 5. Undergraduate GPA ( out of 10 ) 
#6. Research Experience ( either 0 or 1 ) 7. Chance of Admit ( ranging from 0 to 1 )
#We will set the 'Serial No.' column as index, and modify the column names so that they are easier to use later on.
admissions = pd.read_csv("Admission_Predict.csv", index_col='Serial No.')
admissions.columns = admissions.columns.str.lower().str.strip().str.replace(' ', '_')
```

```python
#Suppose we have limited seats available because of which we raise the cutoff value for 'chance of admit' to 75% probability 
#to consider a student eligible for admission.
admissions['admit'] = admissions['chance_of_admit'].apply(lambda x: 1 if x>0.75 else 0)
admissions = admissions.drop('chance_of_admit',axis=1)
print(admissions['admit'].value_counts())
```

```python
#Check for collinearity in between the feature columns.
corrmat = admissions.corr()
import seaborn as sns
import matplotlib.pyplot as plt
fig,axes = plt.subplots(figsize=(10,8))
sns.heatmap(corrmat.abs(), annot=True)
plt.show()
```

```python
#gre_score, toefl_score, cgpa are having high colinearity.
#When 2 features convey almost the same information using 2 different measures or metrics, we don't need to keep both.
#So, we will get their variance and keep only the one with highest variance.
#But before that we will scale the data.
```

```python
minimum = admissions.min(axis=0)
maximum = admissions.max(axis=0)
temp = (admissions - minimum)/(maximum - minimum)
variance = temp.var()
print(variance)
```

```python
#Dropping 'toefl_score' and 'cgpa' because of lower variance.
temp = temp.drop(['toefl_score','cgpa'],axis=1)
```

```python
#Now, a logistic regression model may not converge properly if there are attributes of different scales.
#So, we will use the standardized data to train the model.
```

```python
#We will add an intercept column containing only 1s to the dataset.
intercept = pd.Series([1] * temp.shape[0], index=list(range(1,temp.shape[0]+1)))
temp = pd.concat([intercept, temp], axis=1)
```

```python
#Preparing the training and the test dataset.
np.random.seed(1)
shuffled_index = np.random.permutation(temp.index)
df = temp.reindex(shuffled_index)
train_size = int(0.75*df.shape[0])
train_df = df.iloc[:train_size].copy()
test_df = df.iloc[train_size:].copy()
train_cols = df.columns.drop('admit')
```

```python
#Sigmoid function.
import math
def sigmoid(X, weight):
    z = X.dot(weight)
    # np.exp(x) raises x to the exponential power, ie e^x. e ~= 2.71828
    return 1 / (1 + np.exp(-z))
```

```python
#Gradient calculation.
def gradient_cal(X, prob, y):
    return np.dot(X.T, (prob - y)) / X.shape[0]
```

```python
#Update the weights.
def update_weight(weight, learning_rate, gradient):
    return weight - learning_rate * gradient
```

```python
#Calculate Loss.
def loss(h, y):
    return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
```

```python
#Function to train the weight values.
def train_weights(X, y, weight, lrate=0.1):
    num_iter = 1
    while(num_iter<10000):
        prob = sigmoid(X, weight)
        gradient = gradient_cal(X, prob, y)
        weight = update_weight(weight, lrate, gradient)
        cost = loss(prob, y)
        num_iter += 1
    return weight, num_iter
```

```python
#Training the logistic regression model.
weight = [0] * len(train_cols)
weight, num_iter = train_weights(train_df[train_cols], train_df['admit'], weight)
print(weight)
```

```python
#Applying the model on test dataset.
result = sigmoid(test_df[train_cols], weight)
test_df['pred'] = result.apply(lambda x : 0 if x < 0.5 else 1)
```

```python
#Calculating the true positive rate.
TP = ((test_df['admit']==1) & (test_df['pred']==1)).sum()
FN = ((test_df['admit']==1) & (test_df['pred']==0)).sum()
TPR = TP/(TP+FN)
print(TPR)
```

```python
#Calculating the false positive rate.
TN = ((test_df['admit']==0) & (test_df['pred']==0)).sum()
FP = ((test_df['admit']==0) & (test_df['pred']==1)).sum()
FPR = FP/(FP+TN)
print(FPR)
```
